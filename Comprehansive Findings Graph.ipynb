```python
import matplotlib.pyplot as plt
import numpy as np

def plot_likert_responses(departments, response_counts, colors, title):
    # Initialize ratings
    ratings = np.arange(1, 6)

    # Ensure the lengths match
    if len(departments) != len(response_counts):
        raise ValueError("Length of departments and response_counts must match.")

    # Plotting
    plt.figure(figsize=(10, 6))

    for i, (dept, counts) in enumerate(zip(departments, response_counts)):
        plt.bar(ratings + i*0.15, counts, width=0.15, color=colors[dept], alpha=0.7, label=dept)

    # Add labels and title
    plt.xlabel('Likert Scale')
    plt.ylabel('Number of Responses')
    plt.title(title)

    plt.legend()
    plt.show()

# Example usage
departments = ['Development', 'Professional Service', 'Management']
response_counts = [
    [0, 0, 0, 3, 2],  # Counts for 'Development'
    [0, 0, 0, 0, 1],  # Counts for 'Professional Service'
    [0, 0, 0, 1, 3]   # Counts for 'Management'
]
colors = {'Development': 'blue', 'Professional Service': 'green', 'Management': 'orange'}
title = 'Likert Scale Responses for AI Integration Effectiveness'

plot_likert_responses(departments, response_counts, colors, title)
```

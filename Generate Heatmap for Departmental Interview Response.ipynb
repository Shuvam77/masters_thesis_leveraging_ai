```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def plot_heatmap(questions, ratings, title, ytick_labels, filename='heatmap.png'):
    # Convert ratings to a NumPy array
    ratings = np.array(ratings)

    # Create a heatmap
    fig, ax = plt.subplots(figsize=(12, 8), dpi=300)
    sns.heatmap(
        ratings.T, 
        annot=True, 
        cmap='YlGnBu', 
        xticklabels=questions, 
        yticklabels=ytick_labels, 
        cbar_kws={'label': 'Number of Respondents'},
        linewidths=0.5, 
        linecolor='gray', 
        annot_kws={'size': 10}
    )

    # Adding labels and title
    ax.set_xlabel('Questions', fontsize=12)
    ax.set_ylabel('Ratings', fontsize=12)
    ax.set_title(title, fontsize=14, weight='bold')

    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.yticks(fontsize=10)

    plt.tight_layout()
    plt.savefig(filename, dpi=300)
    plt.show()
```

```python
# Example usage

questions = [
    "AI Impact", 
    "AI Comfort", 
    "AI in Team Coord.", 
    "AI in Decision Making", 
    "AI for Data & Sched.", 
    "AI Effectiveness"
]

ratings = [
    [0, 0, 0, 0, 4],
    [0, 0, 2, 2, 0],
    [0, 0, 0, 1, 3],
    [0, 0, 0, 0, 4],
    [0, 0, 0, 1, 3],
    [0, 0, 0, 2, 2]
]

ytick_labels = [1, 2, 3, 4, 5]

title = 'Survey Responses on Example Department'

plot_heatmap(questions, ratings, title, ytick_labels,'example_heat_map.png')
```
